```python
import pandas as pd

# Load your dataset (replace with your actual file)
#df = pd.read_csv('your_dataset.csv')

# Define the target variable
target_variable = 'Label'

# Initialize an empty list to store sampled dataframes
sampled_dfs = []

# Group by the target variable and sample 10,000 rows from each group
for value in df[target_variable].unique():
    group = df[df[target_variable] == value]
    if len(group) >= 10000:
        sampled_df = group.sample(n=10000, random_state=42)
    else:
        sampled_df = group  # Take all rows if less than 10,000
    sampled_dfs.append(sampled_df)

# Concatenate all the sampled dataframes
subset_df = pd.concat(sampled_dfs, axis=0)

# Shuffle the combined subset to mix the rows
subset_df = subset_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Now, `subset_df` contains a subset of your dataset
print(subset_df)

# Save the subset to a new file if needed
subset_df.to_csv('subset_dataset.csv', index=False)
```

```python
import pandas as pd

# Load your dataset
df = pd.read_csv('subset_dataset.csv')

# Define the split point (e.g., 80% and 20%)
split_point = int(0.8 * len(df))

# Split the dataset
df_part1 = df[:split_point]
df_part2 = df[split_point:]

# Save the parts as CSV files
df_part1.to_csv('dataset_part1.csv', index=False)
df_part2.to_csv('dataset_part2.csv', index=False)

print("Dataset split into two parts and saved as CSV files.")
```
